```python
# Bibliotheken importieren
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os


print("Current working directory:", os.getcwd())

# Daten laden
file_path = '../resources/data_clean/heart_2020_clean.csv'
df = pd.read_csv(file_path)
sns.set(style="whitegrid")
```

```python
# Bivariate Analysis: HeartDisease vs BMI
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='BMI', y='HeartDisease', palette='viridis')
plt.title('Boxplot of HeartDisease vs BMI')
plt.xlabel('BMI')
plt.ylabel('HeartDisease')
plt.show()
```

```python
# Bivariate Analysis: Smoking vs PhysicalHealth und darunter HeartDisease haben 
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='PhysicalHealth', y='Smoking', hue='HeartDisease', palette='viridis')
plt.title('Boxplot of Smoking vs PhysicalHealth')
plt.xlabel('PhysicalHealth')
plt.ylabel('Smoking')
plt.show()
```

```python
# Bivariate Analysis: AgeCategory vs SleepTime
plt.figure(figsize=(6, 3))
sns.boxplot(data=df, x='SleepTime', y='AgeCategory', palette='viridis')
plt.title('Boxplot of AgeCategory vs SleepTime')
plt.xlabel('SleepTime')
plt.ylabel('AgeCategory')
plt.show()
```

```python
# Bivariate Analysis: HeartDisease vs MentalHealth
plt.figure(figsize=(6, 3))
sns.boxplot(data=df, x='MentalHealth', y='HeartDisease', palette='viridis')
plt.title('Boxplot of HeartDisease vs MentalHealth')
plt.xlabel('MentalHealth')
plt.ylabel('HeartDisease')
plt.show()
```

```python
# Bivariate Analysis: Sex vs BMI
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='BMI', y='Sex', hue='HeartDisease',palette='viridis')
plt.title('Boxplot of Sex vs BMI')
plt.xlabel('BMI')
plt.ylabel('Sex')
plt.show()
```

```python
# Bivariate Analysis: HeartDisease vs PhysicalActivity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='HeartDisease', hue='PhysicalActivity', palette='viridis')
plt.title('Countplot of HeartDisease vs PhysicalActivity')
plt.xlabel('HeartDisease')
plt.ylabel('Count')
plt.show()
```

```python
# Bivariate Analysis: Stroke vs Diabetic
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Stroke', hue='Diabetic', palette='viridis')
plt.title('Countplot of Stroke vs Diabetic')
plt.xlabel('Stroke')
plt.ylabel('Count')
plt.show()
```

```python
# Bivariate Analysis: AgeCategory vs MentalHealth
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='AgeCategory', y='MentalHealth', palette='viridis')
plt.title('Boxplot of AgeCategory vs MentalHealth')
plt.xlabel('AgeCategory')
plt.ylabel('MentalHealth')
plt.show()
```

```python
# Bivariate Analysis: BMI vs PhysicalHealth
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='BMI', y='PhysicalHealth', hue='HeartDisease', palette='dark', alpha=0.7)
plt.title('Scatterplot of BMI vs PhysicalHealth')
plt.xlabel('BMI')
plt.ylabel('PhysicalHealth')
plt.show()
```
