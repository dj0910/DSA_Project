```python
# Bibliotheken importieren
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os


print("Current working directory:", os.getcwd())

# Daten laden
file_path = '../resources/data_clean/heart_2020_clean.csv'
data = pd.read_csv(file_path)
sns.set(style="whitegrid")
```

```python
# Univariate Analyse für Spalte 1
column = 'HeartDisease' 
plt.figure(figsize=(6, 3))
if data[column].dtype == 'object':
    sns.countplot(y=column, data=data, order=data[column].value_counts().index, palette='viridis')
    plt.title(f'Verteilung von {column}')
    plt.xlabel('Anzahl')
    plt.ylabel(column)
else:
    sns.histplot(data[column], kde=True, color='blue', binwidth=0.6)
    plt.title(f'Verteilung von {column}')
    plt.xlabel(column)
    plt.ylabel('Dichte')
plt.show()
```
